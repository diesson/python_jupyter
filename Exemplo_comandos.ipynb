```python
# Exemplo String

separador = '------------------------------------'

print("numero: %d" % 10)
print("%d %s %.2f" % (10, 'teste', 1.2555555))
print("{nome} {idade}".format(nome='diesson', idade=23))
print(separador)

# Exemplo tupla
a = (1, 2, 'texto', 3, 3, 4.5, 3, 3)

print("imprimir a tupla: %s" % (a,))
print("seg. elemento: %d" % a[1])
print("onde esta o 1: %d" % a.index(1))
print("onde esta o 'texto': %d" % a.index('texto'))
print("qtd de 3: %d" % a.count(3))

print(separador)
# Exemplo lista

b = [1, 2, 2, 4, (2, 4), [1, 'texto', 3]]
print(b)

b.append(4)
print("add um novo elemento: %s" % (b,))

b[2] = 'texto'
print("mudado o 3 elemento: %s" % (b,))

print("seg. elemento: %d" % b[1])
print("seg. elemento da tupla: %d" % b[4][1])
print("qtd de 2: %d" % b.count(2))
print("onde esta o 4: %d" % b.index(4))

b.insert(0, 5)
print("mudar o prim. elemento: %s" % (b,))

b.reverse()
print("inverter a lista: %s" % (b,))

b = [1, 3, 5, 2, 4]
b.sort()
print("ordenar a lista: %s" % (b,))

print(separador)
# Exemplo dicionario

d = {1: 'Maria', 2: 'Joao', 3: 'Diesson'}

print(d)

print("chave 2: ", d[2])

print('Removido ultimo item: ')
d.popitem()
print(d)

print('Removido 2 item: ')
d.pop(2)
print(d)

print('limpar: ')
d.clear()
print(d)

print(separador)
# Exemplo for

lista = ['Folha', 'Flor', 'Tronco']

for arvore in lista: #for x in range(3)
    print(arvore)
else:
    print(lista)

print(separador)
# Exemplo while

lista = ['Rei', 'Rainha', 'Bispo']

i = 0

while i < len(lista):
    print(lista[i])
    i += 1
else:
    print(lista)

print(separador)


# Exemplo Funcao

def soma(n1, n2=2):
    return n1 + n2


def calc(val1, val2, func):
    return func(val1, val2)


def mul(x, y):
    return x * y


print(soma(1, 10))
print(soma(10))

d = {'n1': 1, 'n2': 4}
print(soma(**d))

opt = 'mul'

if opt == 'soma':
    print(calc(10, 7, soma))
elif opt == 'mul':
    print(calc(10, 7, mul))
```
